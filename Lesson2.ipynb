```python
guess = int(input("Guess a number between and 1 and 10: "))
```

### Operators

### Math Opeators:
* \+ plus           e.g. =>  3 + 2 = 5
* \- minus e.g. =>  3 - 2 = 1
* \/ divide e.g. =>  3 / 2 = 1.5
* \// floor division  e.g. =>  3 // 2 = 1
* \%  modulus  e.g. =>  3 % 2 = 1
* \** to power e.g. =>  3 ** 2 = 9


### Conditional Operators:

*   Less Than < e.g. =>  3 < 2 = **False**
*   Less than or equal to <= e.g. =>  3 <= 2 = *False*
*   Greate Than > e.g. =>  3 > 2 = **True**
* Greater than or equal to >= e.g. =>  3 => 2 = **True**
* Equal to == e.g. =>  3 == 2 = *False*
* Not Equal to != e.g. =>  3 != 2 = **True**


### Logical Operators:
* IF, ELIF, ELSE
* and (True and True) = True, (True and False) = False
* or  (True or True) = True, (True or False) = True
* not (True) = False e.g. Not True = False
* xor (True xor False) = True

## Loops 

* Ranges - creates range of **values**
* For Loops -  for a speified length
* While Loops - while the condition is true
* Do Loops - do until when condition is true

```python
numbers = 100

for i in range(numbers):
  if i == 50: print(i)
```

```python
numbers = 11
for i in range(2,numbers,2):
  if i%2==0: print(i, " is an even number")
  else: print(i, "is an odd number") 
```
